```python
dog_img=Image.open("dog.jpeg")
dog_img.show()
```

5.图片剪裁

```python
zone=(800,200,1500,800)
dog_img_crop=dog_img.crop(zone)
# dog_img_crop.show()
dog_img_crop.save("dog_crop.jpg")
```

6.图片像素提取

```python
dog_img.getpixel((0,0))
```

7.Greyscale:

```python
dog_img_bw=dog_img.convert("L")
dog_img_bw.save("dog_img_bw.jpg")
# dog_img_bw.show()
dog_img_bw.getpixel((0,0))
```

8.黑白转换

```python
dog_img_true_bw=dog_img.convert("1")
# dog_img_true_bw.show()
# dog_img_bw.show()
dog_img_true_bw.getpixel((0,0))
```

```python
resizing_dog_img=dog_img.resize((8000,5000))
# resizing_dog_img.show()
```

9.图像旋转(逆时针)

```python
dog_new_img=dog_img.rotate(180)
# dog_new_img.show()
```

10.亮度调整（brightness）

```python
from PIL import ImageEnhance
enhanced_brightness_dog_img=ImageEnhance.Brightness(dog_img).enhance(2)
enhanced_brightness_dog_img.getpixel((0,0))
```

```python
dog_img.getpixel((120,150))
```

11.对比度（Contrast）

```python
enhanced_contrast_dog_img=ImageEnhance.Contrast(dog_img).enhance(2)
enhanced_contrast_dog_img.getpixel((0,0))
```

12.像素点提取或者替换

```python
dog_img.getpixel((0,0))
```

```python
dog_img.putpixel((0,0),(188,188,188))

enhanced_contrast_dog_img=ImageEnhance.Contrast(dog_img).enhance(2)
enhanced_contrast_dog_img.getpixel((0,0))
```

13.scikit-image

I/O IO (input and output)

```python
from skimage import io
dog_img=io.imread("dog.jpeg")
io.imshow(dog_img)
io.show()
```

```python
io.imsave("new_dog.jpeg",dog_img)
```

jpg         GIF(Graphics Interchange Format)     PNG(Portable Network Graphics)

```python
from skimage import data,color
io.imshow(data.coffee())
io.show()
```

```python
coffee_img=data.coffee()
grey_coffee_img=color.rgb2grey(coffee_img)
io.imshow(grey_coffee_img)
io.show()
```
