```python
import sympy as sym
```

```python
sym.init_printing()
```

```python
x,y = sym.symbols("x y")
```

```python
f = x**2*y**3+sym.cos(y)+sym.sin(x)
```

```python
sym.Derivative(f,x)
```

```python
sym.diff(f,x)
```

```python
sym.diff(f,y)
```

```python
%matplotlib inline

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection = "3d")
```

```python
%matplotlib notebook
def func(x,y):
     return x**2*y+np.cos(y)+np.sin(x)

x = np.linspace(-888,888,8)
y = np.linspace(-888,888,8)
```

```python
x
```

```python
y
```

```python
X, Y =  np.meshgrid(x,y)
X
```

```python
Y
```

```python
Z = func(X,Y)
```

```python
Z
```

```python
fig =plt.figure()
ax= plt.axes(projection = "3d")
ax.contour3D(X,Y,Z,50,cmap="binary")
```

```python
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap="viridis",edgecolor="none")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```

$function : f(x,y) \\
Gradient Vector : \nabla f(x,y) = \begin{bmatrix}
\frac{\partial f(x,y)}{\partial x}\\
\frac{\partial f(x,y)}{\partial y}
\end{bmatrix}=\begin{bmatrix} f_x\\f_y\end{bmatrix}\\
Grandient magnitude= |\nabla f(x,y)|=\sqrt{f_x^2+f_y^2} \\
Grawdient Direction = \theta =tan^-1 \frac{f_x}{f_y}
$
