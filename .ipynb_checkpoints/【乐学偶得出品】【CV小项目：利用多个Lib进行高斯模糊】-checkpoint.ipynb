```python
from PIL import Image
from PIL import ImageFilter
img = Image.open("6.Gaussian_Noise.png")
blurred_img = img.filter(ImageFilter.GaussianBlur(5))
blurred_img.show()
img.show()
```

```python
from skimage import io
from skimage import filters
img = io.imread("6.Gaussian_Noise.png")
blurred_img = filters.gaussian(img,sigma=1)
io.imshow(blurred_img)
io.show()
```
