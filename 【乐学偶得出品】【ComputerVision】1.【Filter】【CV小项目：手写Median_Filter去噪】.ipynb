```python
from PIL import Image
```

```python
path = "salt_and_peper.png"
```

```python
img = Image.open(path)
```

```python
img
```

```python
filter_mask = [(0,0)] * 9
```

```python
filter_mask
```

```python
width=img.width
```

```python
height=img.height
```

```python
newimg = Image.new("L",(width,height),"white")
```

```python
newimg.show()
```

```python
for i in range(1,width-1):
    for j in range(1,height-1):
        filter_mask[0] = img.getpixel((i-1,j-1))
        filter_mask[1] = img.getpixel((i-1,j))
        filter_mask[2] = img.getpixel((i-1,j+1))
        filter_mask[3] = img.getpixel((i,j-1))
        filter_mask[4] = img.getpixel((i,j))
        filter_mask[5] = img.getpixel((i,j+1))
        filter_mask[6] = img.getpixel((i+1,j-1))
        filter_mask[7] = img.getpixel((i+1,j))
        filter_mask[8] = img.getpixel((i+1,j+1))
        filter_mask.sort()
        newimg.putpixel((i,j),(filter_mask[4]))
```

```python
newimg.show()
```
